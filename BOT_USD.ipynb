```python
import pandas as pd
CY = 'AUD'
rate = pd.read_html(f'https://rate.bot.com.tw/xrt/quote/l6m/{CY}')
rate = rate[0]
rate = rate.iloc[:,0:6]
rate.columns = [u'掛牌日期',u'幣別',u'現金賣出',u'現金買入',u'即期買入',u'即期賣出']
rate['幣別'] = rate['幣別'].str.extract('\((\w+)\)')
rate['即期價差'] = rate['即期買入']-rate['即期賣出']
rate['即期價差%'] = rate['即期價差']/rate['即期賣出']*100
rate
```

```python
rate.info()
```

```python
rate['掛牌日期'] = pd.to_datetime(rate['掛牌日期'])
```

```python
import sqlite3
with sqlite3.connect('curncy.sqlite') as db:
    rate.to_sql('curncy',con = db, if_exists='append')
```

```python
with sqlite3.connect('curncy.sqlite') as db:
    df = pd.read_sql("select * from curncy WHERE curncy.'即期買入'> 22.2", con=db)
df
```
