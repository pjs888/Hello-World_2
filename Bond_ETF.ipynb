```
import requests
import pandas as pd
from datetime import datetime
from bs4 import BeautifulSoup
```

```
# 抓取OTC之債券ETF基本資料代號與名稱
Bond_ETF_OTC = pd.read_json('https://www.tpex.org.tw/web/etf/etf_specification_bond_result.php?l=zh-tw&_=1591756038948%20Request%20Method:%20GET')
Bond_ETF_OTC = Bond_ETF_OTC['aaData']
```

```
# 抓取Moneydj之債券ETF基本資料
def get_Moneydj_Data(ID):
    ETF_Basic_Data={}
    url = 'https://www.moneydj.com/ETF/X/Basic/Basic0004.xdjhtm?etfid=' + str(ID) +'.TW'
    r = requests.get(url)
    r.raise_for_status()
    r.encoding = 'urf-8'
    soup = BeautifulSoup(r.text, 'html.parser')
    Attrs = [soup.find_all("th")[i].text for i in range(len(soup.find_all("th")))] 
    for Attr in Attrs:
        ETF_Basic_Data[Attr] = soup.find("th", text=Attr).find_next_sibling("td").text.strip()
    return ETF_Basic_Data
```

```
# 抓取全部債券ETF之基本資料
ETF_DATA = {}
for i in range(len(Bond_ETF_OTC)):
    ID = Bond_ETF_OTC[i][0]
    ETF_DATA[ID] = get_Moneydj_Data(ID)
```

```
Index = []
for ETF in Bond_ETF_OTC:
    X = ETF[0]
    X = ETF_DATA[X]
    X = X['追蹤指數']
    Index.append(X)    
Index.sort()
Index
```
